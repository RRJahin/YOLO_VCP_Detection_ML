```
import os
import shutil
from sklearn.model_selection import train_test_split
```

```
os.listdir('/content/drive/MyDrive/Colab Notebooks/ML Project/kitti_single')
DATA_DIR = '/content/drive/MyDrive/Colab Notebooks/ML Project/kitti_single'
```

```
LABELS = DATA_DIR + "/training/yolo-labels/"
IMAGES = DATA_DIR + "/training/image_2"
OUT = DATA_DIR + "/data"
```

```
labels = os.listdir(LABELS)
print(len(labels))
```

```
if os.path.exists(OUT):
  shutil.rmtree(OUT)

os.makedirs(OUT)
os.makedirs(OUT + "/train")
os.makedirs(OUT + "/valid")
```

```
TOTAL_SIZE = 3000
train,valid = train_test_split(labels[:TOTAL_SIZE],test_size=0.2, random_state=64)
print("#Training Examples :",len(train))
print("#Validation Examples :",len(valid))
```

```
count,total = 1,len(train)
for label in train:
    print(f"train/ {count}/{total}")
    count += 1

    label_file = f"/{label}"
    image_file = f"/{label.split('.')[0]}.png"
    shutil.copyfile(LABELS + label_file,OUT + "/train" + label_file)
    shutil.copyfile(IMAGES + image_file,OUT + "/train" + image_file)

count,total = 1,len(valid)
for label in valid:
    print(f"valid/ {count}/{total}")
    count += 1
    label_file = f"/{label}"
    image_file = f"/{label.split('.')[0]}.png"
    shutil.copyfile(LABELS + label_file,OUT + "/valid/" + label_file)
    shutil.copyfile(IMAGES + image_file,OUT + "/valid" + image_file)
```
